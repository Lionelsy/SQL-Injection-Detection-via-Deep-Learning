```python
df = pd.read_csv("../archive/sqliv2.csv",encoding='utf-16')
```

# Pre-Process

```python
df.shape
```

```python
data = df.drop_duplicates('Sentence')
```

```python
data.shape
```

```python
data.to_csv('../data/all.csv', index=None)
```

```python
attack = data[data['Label']==1][1:]
attack = attack.sample(frac=1).reset_index(drop=True)
```

```python
normal = data[data['Label']==0]
normal = normal.sample(frac=1).reset_index(drop=True)
```

```python
r1 = 0.8
r2 = 0.9
```

```python
a1 = attack[:int(attack.shape[0]*r1)]
a2 = attack[int(attack.shape[0]*r1):int(attack.shape[0]*r2)]
a3 = attack[int(attack.shape[0]*r2):]
```

```python
a3
```

```python
normal
```

```python
n1 = normal[:int(normal.shape[0]*r1)]
n2 = normal[int(normal.shape[0]*r1):int(normal.shape[0]*r2)]
n3 = normal[int(normal.shape[0]*r2):]
```

```python
n1
```

```python
train = a1.append(n1)
val = a2.append(n2)
test = a3.append(n3)
```

```python
train = train.sample(frac=1).reset_index(drop=True)
val = val.sample(frac=1).reset_index(drop=True)
test = test.sample(frac=1).reset_index(drop=True)
```
